## Challenge 1
We want to estimate the average size of the men of a country with a confidence level of 80%. Assuming that the standard deviation of the sizes in the population is 4, get the confidence interval with a sample of men selected randomly, whose heights are:

````
heights = [167, 167, 168, 168, 168, 169, 171, 172, 173, 175, 175, 175, 177, 182, 195]
````

**Hint**: function `stats.norm.interval` from `scipy` can help you get through this exercise. 

`````python
#your code here
heights = np.array([167, 167, 168, 168, 169, 169, 171, 172, 173, 175, 175, 175, 177, 182, 195])

#calculating the mean
mean_sample = heights.mean()

#Calculating the std of the sample 
std_pop=4
std_sample=std_pop/(len(heights)**0.5)

probability_heights_pop = norm(mean_sample, std_pop)
probability_heights = norm(mean_sample, std_sample)
plt.plot(heights,probability_heights_pop.pdf(heights))

alpha=0.8
interval_avg_heights=probability_heights_pop.interval(alpha)
interval_avg_heights

#Calculating the std of the sample
`````

`````python
# your code here
heights = np.array([167, 167, 168, 168, 169, 169, 171, 172, 173, 175, 175, 175, 177, 182, 195])
#We can not talk about the number of people because we are using the mano f a sample 
#So the interval confidence is not about population but mean 
#DF – Degree of Freedom = IF you think about at least its 0 Dimension , only 1 sample 
#N – Lengh of the sangle

# T DISTRIBUTION when sample lenght is less than 30, we use the T distribution.
# If greater, use the normal distribution
#x=np.arange(150,190)
heights = np.array([167, 167, 168, 168, 168, 169, 171, 172, 173, 175, 175, 175, 177, 182, 195])
std_sample=heights.std()
mean_sample= heights.mean()
degree_freedom= len(heights) - 1
t_distribution_heights = t(degree_freedom,mean_sample,std_sample)
plt.plot(heights,t_distribution_heights.pdf(heights))
alpha=0.8
interval_t_distrib=t_distribution_heights.interval(alpha)
interval_t_distrib
`````
